```python
import pandas as pd
import numpy as np

np.random.seed(42)

rows = 10000

deployment_days = [
    "Monday","Tuesday","Wednesday",
    "Thursday","Friday","Saturday","Sunday"
]

change_types = [
    "Application",
    "Database",
    "Infrastructure",
    "Network"
]

data = pd.DataFrame({

    "systems_impacted":
        np.random.randint(1,15,rows),

    "num_files_changed":
        np.random.randint(1,100,rows),

    "deployment_hour":
        np.random.randint(0,24,rows),

    "deployment_day":
        np.random.choice(deployment_days,rows),

    "engineer_experience":
        np.random.randint(1,15,rows),

    "historical_failures":
        np.random.randint(0,6,rows),

    "dependency_count":
        np.random.randint(0,12,rows),

    "rollback_required":
        np.random.randint(0,2,rows),

    "test_coverage":
        np.random.randint(40,101,rows),

    "change_type":
        np.random.choice(change_types,rows)
})
```

```python
risk_score = (

      data["dependency_count"] * 2
    + data["historical_failures"] * 3
    + data["systems_impacted"] * 1.5
    + data["rollback_required"] * 5
    + (100 - data["test_coverage"]) * 0.1

)

risk_score += np.where(
    data["deployment_hour"] > 21,
    5,
    0
)

risk_score += np.where(
    data["deployment_day"]=="Friday",
    4,
    0
)

risk_score += np.random.normal(0,3,rows)
```

```python
threshold = np.percentile(risk_score,65)

data["change_failed"] = (
    risk_score > threshold
).astype(int)
```

```python
data["change_failed"].value_counts()
```

```python
data.to_csv(
    "deployment_data.csv",
    index=False
)

print("saved")
```
