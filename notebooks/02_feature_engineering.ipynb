```python
import pandas as pd

df = pd.read_csv("../data/deployment_data.csv")

df.head()
```

```python
df["complexity_score"] = (
    df["systems_impacted"]
    * df["dependency_count"]
)
```

```python
df["after_hours"] = (
    df["deployment_hour"] > 18
).astype(int)
```

```python
df["weekend_deployment"] = (
    df["deployment_day"]
      .isin(["Saturday", "Sunday"])
).astype(int)
```

```python
change_type_encoded = pd.get_dummies(
    df["change_type"],
    prefix="change_type"
)

df = pd.concat(
    [df, change_type_encoded],
    axis=1
)
```

```python
df.to_csv(
    "../data/deployment_data_engineered.csv",
    index=False
)

print("Feature engineering completed.")
```
